```python
planes = pd.read_csv("plane-data.csv").dropna()
planes.shape
```

```python
planes.head()
```

```python
df_planes = pd.merge(data_2002, planes[['tailnum', 'year']], on='tailnum', how='inner')
df_planes.head()
```

```python
df_planes.drop(index=(df_planes[df_planes['year'] == 'None']).index, inplace=True)
df_planes.shape
```

```python
df_planes['year'] = df_planes['year'].astype(int)
```

```python
df_planes["Plane_age"] = df_planes['Year'] - df_planes['year']
```

```python
df_planes.head()
```

```python
from scipy.stats import pearsonr
pearsonr(df_planes['ArrDelay'], df_planes['Plane_age'])
```

```python
df_planes["Plane_age"] = df_planes['Year'] - df_planes['year']
df_planes.tail()
```

```python
df_planes.groupby("Plane_age")['ArrDelay'].mean()
```
